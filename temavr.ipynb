```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/GlobalTemperatures.csv',parse_dates=['date'])
df.fillna(method='ffill')
df['year']=df['date'].dt.year
df['month']=df['date'].dt.month
```

```python
YAT = pd.DataFrame(columns=['year','avr'])
YAT['avr']=df.groupby('year')['LAT'].mean()
```

```python
YAT['year']=list(range(1750, 2016))
```

```python
co=pd.read_csv('./data/CO2 emission by countries.csv')
```

```python
co2 = pd.DataFrame(columns=['year','avr'])
co2['avr']=co.groupby('year')['CO2'].mean()
co2['year']=list(range(1750, 2021))
```

```python
co2=co2[co2['year']<=2015]
```

```python
co2
```

```python
plt.plot(YAT['year'],YAT['avr'])
plt.show()
plt.plot(co2['year'],co2['avr'])
plt.show()
```

```python
fig, ax1 = plt.subplots()
ax1.plot(YAT['year'], YAT['avr'], color = 'red', alpha = 0.5)

ax2 = ax1.twinx()
ax2.plot(YAT['year'], co2['avr'], color = 'blue', alpha = 0.5)
ax2.set_ylim(-8000000000, 8000000000)

plt.title('Temperature changes with CO2', fontsize = 10, color ='red')

plt.show()
```
